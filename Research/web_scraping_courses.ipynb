```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://courses.illinois.edu/schedule/2021/fall")
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
soup.find_all("a")
```

```python
majors_ = [] 
for row in soup.findAll('tr')[1:]:
    col = row.findAll('td')
    majors_.append(col[0].getText())
```

```python
cleaned_ = []
for item in majors_:
    cleaned_.append(item.strip().replace("\n", ""))
```

```python
cleaned_
```

```python
for item in cleaned_:
  new_url = "https://courses.illinois.edu/schedule/2021/fall/" + item  
  print(new_url) 
```

```python
majors_courses = [] 
for item in cleaned_:
  new_url = "https://courses.illinois.edu/schedule/2021/fall/" + item
  major_page = requests.get(new_url)
  majoir_soup = BeautifulSoup(major_page.content, 'html.parser')
  for row in majoir_soup.findAll('tr')[1:]:
      col = row.findAll('td')
      majors_courses.append(col[0].getText())

for item in majors_courses:
    cleaned_.append(item.strip().replace("\n", ""))
print(major_courses)
```
